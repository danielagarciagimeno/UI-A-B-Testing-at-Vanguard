## Time Spent on Each Step

```python
# Crear una nueva columna con la marca de tiempo de la siguiente etapa para cada visita
df_control.loc[:, 'next_date_time'] = df_control.groupby('visit_id')['date_time'].shift(-1)

# Calcular el tiempo dedicado en cada etapa (en segundos)
df_control.loc[:, 'time_spent'] = (df_control['next_date_time'] - df_control['date_time']).dt.total_seconds()

# Calcular el promedio de tiempo dedicado por etapa
time_spent_avg_control = df_control.groupby('process_step')['time_spent'].mean()

# Mostrar los resultados
print("Tiempo promedio dedicado por etapa en el grupo de control:")
print(time_spent_avg_control)
```

```python
# Crear una nueva columna con la marca de tiempo de la siguiente etapa para cada visita
df_test.loc[:, 'next_date_time'] = df_test.groupby('visit_id')['date_time'].shift(-1)

# Calcular el tiempo dedicado en cada etapa (en segundos)
df_test.loc[:, 'time_spent'] = (df_test['next_date_time'] - df_test['date_time']).dt.total_seconds()

# Calcular el promedio de tiempo dedicado por etapa
time_spent_avg_test = df_test.groupby('process_step')['time_spent'].mean()

# Mostrar los resultados
print("Tiempo promedio dedicado por etapa en el grupo de test:")
print(time_spent_avg_test)
```

## Error Rate

Passos para calcular a taxa de erros:
1. Identificar quando um usuário volta para uma etapa anterior.
2. Contar o número de "erros" por client_id ou visit_id.
3. Calcular a taxa de erro: A taxa de erro pode ser calculada como a razão entre o número de erros e o número total de visitas ou clientes.

```python
# Ordenar los dataframes por client_id, visit_id y date_time
df_control = df_control.sort_values(by=['client_id', 'visit_id', 'date_time'])
df_test = df_test.sort_values(by=['client_id', 'visit_id', 'date_time'])

# Crear una nueva columna con la siguiente etapa para cada visita
df_control['next_process_step'] = df_control.groupby('visit_id')['process_step'].shift(-1)
df_test['next_process_step'] = df_test.groupby('visit_id')['process_step'].shift(-1)

# Definir el orden de las etapas del proceso
process_steps_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Identificar retrocesos (si la siguiente etapa es anterior en el proceso)
df_control['step_back'] = df_control.apply(lambda row: process_steps_order.index(row['process_step']) > process_steps_order.index(row['next_process_step']) if pd.notnull(row['next_process_step']) else False, axis=1)
df_test['step_back'] = df_test.apply(lambda row: process_steps_order.index(row['process_step']) > process_steps_order.index(row['next_process_step']) if pd.notnull(row['next_process_step']) else False, axis=1)
```

```python
# Contabilizar los errores para el grupo de control
users_with_errors_control = df_control[df_control['step_back'] == True]['client_id'].unique()
num_users_with_errors_control = len(users_with_errors_control)
total_users_control = df_control['client_id'].nunique()
error_rate_control = num_users_with_errors_control / total_users_control

# Contabilizar los errores para el grupo de prueba
users_with_errors_test = df_test[df_test['step_back'] == True]['client_id'].unique()
num_users_with_errors_test = len(users_with_errors_test)
total_users_test = df_test['client_id'].nunique()
error_rate_test = num_users_with_errors_test / total_users_test

# Mostrar las tasas de error para ambos grupos
print(f'Tasa de error del grupo de control: {error_rate_control:.2%}')
print(f'Tasa de error del grupo de prueba: {error_rate_test:.2%}')
```
