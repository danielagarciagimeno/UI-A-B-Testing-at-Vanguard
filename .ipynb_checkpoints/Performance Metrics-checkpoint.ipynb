## Calculo del Timpo total para completar el proceso en cada grupo:


```python
import pandas as pd

# Asegurar que la columna 'date_time' está en el formato datetime
df_test['date_time'] = pd.to_datetime(df_test['date_time'])
df_control['date_time'] = pd.to_datetime(df_control['date_time'])

# Función para calcular el tiempo promedio por cliente
def calcular_tiempo_por_cliente(df):
    # Ordenar el DataFrame por client_id, visit_id y fecha/hora
    df = df.sort_values(by=['client_id', 'visit_id', 'date_time'])

    # Identificar el primer 'start' y el último 'confirm' para cada cliente y sesión
    df_start = df[df['process_step'] == 'start'].groupby(['client_id', 'visit_id'])['date_time'].min()
    df_confirm = df[df['process_step'] == 'confirm'].groupby(['client_id', 'visit_id'])['date_time'].max()

    # Combinar los tiempos de inicio y confirmación en un DataFrame único
    tiempo_por_sesion = pd.DataFrame({'start': df_start, 'confirm': df_confirm})
    
    # Calcular el tiempo total por sesión (en segundos)
    tiempo_por_sesion['time_spent'] = (tiempo_por_sesion['confirm'] - tiempo_por_sesion['start']).dt.total_seconds()

    # Eliminar sesiones con valores NaN (clientes sin 'start' o 'confirm')
    tiempo_por_sesion = tiempo_por_sesion.dropna()

    # Calcular el tiempo promedio por cliente
    tiempo_promedio_cliente = tiempo_por_sesion.groupby('client_id')['time_spent'].mean().mean()
    
    return tiempo_promedio_cliente

# Calcular el tiempo promedio por cliente para los grupos de Test y Control
tiempo_promedio_test = calcular_tiempo_por_cliente(df_test)
tiempo_promedio_control = calcular_tiempo_por_cliente(df_control)

# Mostrar los resultados
print(f"Tiempo promedio por cliente en el grupo de Test: {tiempo_promedio_test:.2f} segundos")
print(f"Tiempo promedio por cliente en el grupo de Control: {tiempo_promedio_control:.2f} segundos")
```

```python
# Exportar los resultados a un archivo CSV
data = {
    'grupo': ['Test', 'Control'],
    'tiempo_promedio': [tiempo_promedio_test, tiempo_promedio_control]
}
df_tiempo_promedio = pd.DataFrame(data)

# Exportar a un archivo CSV
df_tiempo_promedio.to_csv('tiempo_promedio_total.csv', index=False)

print("Datos exportados a 'tiempo_promedio_total.csv'")
```

## Calculo de la Error Rate:

Pasos para calcular la tasa de errores:

1. Identificar cuando un usuario vuelve a una etapa anterior.

2. Contar el número de "errores" por client_id o visit_id.

3. Calcular la tasa de error: La tasa de error se puede calcular como la relación entre el número de errores y el número total de visitas o clientes.

```python
# Ordenar los dataframes por client_id, visit_id y date_time
df_control = df_control.sort_values(by=['client_id', 'visit_id', 'date_time'])
df_test = df_test.sort_values(by=['client_id', 'visit_id', 'date_time'])

# Crear una nueva columna con la siguiente etapa para cada visita
df_control['next_process_step'] = df_control.groupby('visit_id')['process_step'].shift(-1)
df_test['next_process_step'] = df_test.groupby('visit_id')['process_step'].shift(-1)

# Definir el orden de las etapas del proceso
process_steps_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Identificar retrocesos (si la siguiente etapa es anterior en el proceso)
df_control['step_back'] = df_control.apply(lambda row: process_steps_order.index(row['process_step']) > process_steps_order.index(row['next_process_step']) if pd.notnull(row['next_process_step']) else False, axis=1)
df_test['step_back'] = df_test.apply(lambda row: process_steps_order.index(row['process_step']) > process_steps_order.index(row['next_process_step']) if pd.notnull(row['next_process_step']) else False, axis=1)
```

```python
# Contabilizar los errores para el grupo de control
users_with_errors_control = df_control[df_control['step_back'] == True]['client_id'].unique()
num_users_with_errors_control = len(users_with_errors_control)
total_users_control = df_control['client_id'].nunique()
error_rate_control = num_users_with_errors_control / total_users_control

# Contabilizar los errores para el grupo de prueba
users_with_errors_test = df_test[df_test['step_back'] == True]['client_id'].unique()
num_users_with_errors_test = len(users_with_errors_test)
total_users_test = df_test['client_id'].nunique()
error_rate_test = num_users_with_errors_test / total_users_test

# Mostrar las tasas de error para ambos grupos
print(f'Tasa de error del grupo de control: {error_rate_control:.2%}')
print(f'Tasa de error del grupo de test: {error_rate_test:.2%}')
```

```python
import matplotlib.pyplot as plt

# Datos para las tasas de error
tasa_error = {
    'grupo': ['Control', 'Test'],
    'tasa_error': [26.26, 33.57]
}

# Crear DataFrame
df_tasa_error = pd.DataFrame(tasa_error)

# Crear el gráfico de barras
plt.figure(figsize=(6, 3))
plt.bar(df_tasa_error['grupo'], df_tasa_error['tasa_error'], color=['red', 'blue'])

# Añadir título y etiquetas
plt.title('Comparación de la Tasa de Error entre los Grupos de Control y Test', fontsize=14)
plt.xlabel('Grupo', fontsize=10)
plt.ylabel('Tasa de Error (%)', fontsize=10)

# Mostrar el valor de la tasa de error en la parte superior de cada barra
for i, v in enumerate(df_tasa_error['tasa_error']):
    plt.text(i, v + 1, f'{v}%', ha='center', fontsize=10)

# Mostrar el gráfico
plt.show()
```

```python
# Asegúrese de que la columna 'date_time' está en formato datetime
df_control['date_time'] = pd.to_datetime(df_control['date_time'])
df_test['date_time'] = pd.to_datetime(df_test['date_time'])

# Eliminar duplicados, considerando solo la primera ocurrencia de "confirm" por cliente
df_control_unique = df_control[df_control['process_step'] == 'confirm'].drop_duplicates(subset='client_id', keep='first')
df_test_unique = df_test[df_test['process_step'] == 'confirm'].drop_duplicates(subset='client_id', keep='first')

# Crear la columna 'completed' para los clientes con al menos un "confirm"
df_control_unique['completed'] = 1
df_test_unique['completed'] = 1

# Para los clientes sin "confirm", marcar como 0
df_control_no_confirm = df_control[~df_control['client_id'].isin(df_control_unique['client_id'])]
df_test_no_confirm = df_test[~df_test['client_id'].isin(df_test_unique['client_id'])]

# Usar .loc[] para evitar el SettingWithCopyWarning
df_control_no_confirm.loc[:, 'completed'] = 0
df_test_no_confirm.loc[:, 'completed'] = 0

# Concatenar los DataFrames
df_control_final = pd.concat([df_control_unique[['client_id', 'completed']], df_control_no_confirm[['client_id', 'completed']]])
df_test_final = pd.concat([df_test_unique[['client_id', 'completed']], df_test_no_confirm[['client_id', 'completed']]])

# Verificar los resultados finales
print("Clientes completados en el grupo de control:")
print(df_control_final.drop_duplicates(subset='client_id'))

print("Clientes completados en el grupo de prueba:")
print(df_test_final.drop_duplicates(subset='client_id'))
```

#### Hipótesis Nula (H₀): Las tasas de completación son iguales entre los grupos de control y prueba (no hay diferencia significativa).

#### Hipótesis Alternativa (H₁): Las tasas de completación son diferentes entre los grupos de control y prueba (hay una diferencia significativa).

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.proportion import proportions_ztest

# Calcular el número de éxitos (clientes completados) y el total de observaciones para cada grupo
exitos_control = df_control_final['completed'].sum()
n_control = df_control_final['completed'].count()

exitos_test = df_test_final['completed'].sum()
n_test = df_test_final['completed'].count()

# Realizar la prueba Z para dos proporciones
conteo = [exitos_test, exitos_control]
nobs = [n_test, n_control]

# Prueba unilateral
estadistica, valor_p = proportions_ztest(conteo, nobs, alternative='larger')  # 'larger' significa que la prueba es unilateral

# Mostrar los resultados
print(f"Estadística de la prueba Z: {estadistica:.10f}")
print(f"Valor-p: {valor_p:.30f}")

# Análisis del resultado
alpha = 0.05  # Nivel de significancia
if valor_p < alpha:
    print("Rechazamos la hipótesis nula: Existe una diferencia significativa entre los grupos.")
else:
    print("No rechazamos la hipótesis nula: No existe una diferencia significativa entre los grupos.")
```

#### Como el valor-p es menor que 0,05, esto significa que la diferencia entre las tasas de finalización de los dos grupos es estadísticamente significativa. Es decir, podemos rechazar la hipótesis nula y concluir que el nuevo diseño tiene un efecto positivo en la tasa de finalización.

```python
import pandas as pd

# Datos de los resultados de la prueba Z
data = {
    'Estadística Z': [estadistica],
    'Valor-p': [valor_p],
    'Resultado': ['Rechazamos la hipótesis nula' if valor_p < alpha else 'No rechazamos la hipótesis nula']
}

# Crear DataFrame
df_resultados = pd.DataFrame(data)

# Exportar a un archivo CSV
df_resultados.to_csv('resultados_prueba_z_completion_rate.csv', index=False)
```

## Hypothesis Testing - Completion Rate with a Cost-Effectiveness Threshold:


```python
# Calcular el aumento observado
observed_increase = ((completion_rate_test - completion_rate_control) / completion_rate_control) * 100

# Umbral mínimo de aumento definido por la empresa
threshold = 5  # Aumento del 5%

# Verificar si el aumento observado es suficiente
if observed_increase >= threshold:
    print(f"Análisis de Costo-Beneficio: El aumento en la tasa de finalización ({observed_increase:.2f}%) es suficiente para justificar el nuevo diseño.")
else:
    print(f"Análisis de Costo-Beneficio: El aumento en la tasa de finalización ({observed_increase:.2f}%) no es suficiente para justificar el nuevo diseño.")
```

## Hypothesis Testing - Error Rates:

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# número de errores en el grupo de control
errors_control = 23532 * 0.2626  # Por ejemplo, 26.26% de errores en el grupo de control

# número de errores en el grupo de prueba
errors_test = 26968 * 0.3357  # Por ejemplo, 33.57% de errores en el grupo de prueba

# total de clientes en cada grupo
n_control = 23532  # Número total de clientes en el grupo de control
n_test = 26968  # Número total de clientes en el grupo de prueba

# Paso 1: Realizar la prueba de proporción (proportions_ztest)
count = [errors_test, errors_control]  # número de errores en los dos grupos
nobs = [n_test, n_control]  # número de observaciones en los dos grupos

# Prueba unilateral (usando 'larger' para ver si el grupo de prueba tiene mayor tasa de error)
stat, p_value = proportions_ztest(count, nobs, alternative='larger')

# Paso 2: Mostrar los resultados
print(f"Estadística de la prueba Z: {stat:.5f}")
print(f"Valor-p: {p_value:.100f}")

# Paso 3: Análisis del resultado
alpha = 0.05  # Nivel de significancia
if p_value < alpha:
    print("Rechazamos la hipótesis nula: Existe una diferencia significativa en las tasas de error entre los grupos.")
else:
    print("No podemos rechazar la hipótesis nula: No existe una diferencia significativa en las tasas de error entre los grupos.")
```

```python
import pandas as pd

# Crear DataFrame con los resultados
data = {
    'Estadística Z': [stat],
    'Valor-p': [p_value],
    'Resultado': ['Rechazamos la hipótesis nula' if p_value < alpha else 'No podemos rechazar la hipótesis nula']
}
df_resultados3 = pd.DataFrame(data)

# Exportar a un archivo CSV
df_resultados3.to_csv('resultados_prueba_z_tasa_error.csv', index=False)

print("Datos exportados a 'resultados_prueba_z_tasa_error.csv'")
```

```python
import scipy.stats as st

# Paso 1: Calcular la prueba t para dos muestras independientes
stat, p_value = st.ttest_ind(tiempo_promedio_control, tiempo_promedio_test, equal_var=False)

# Paso 2: Mostrar los resultados
print(f"Estadística t: {stat:.5f}")
print(f"Valor-p: {p_value:.5f}")

# Paso 3: Análisis del resultado
alpha = 0.05  # Nivel de significancia
if p_value < alpha:
    print("Rechazamos la hipótesis nula: Existe una diferencia significativa entre los tiempos promedio de los grupos.")
else:
    print("No podemos rechazar la hipótesis nula: No existe una diferencia significativa entre los tiempos promedio de los grupos.")
```

```python
from scipy.stats import mannwhitneyu

# Prueba de Mann-Whitney U
stat, p_value = mannwhitneyu(tiempo_promedio_control, tiempo_promedio_test, alternative='two-sided')

print(f"Estadística U: {stat:.5f}")
print(f"Valor-p: {p_value:.5f}")

# Análisis del resultado
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula: Existe una diferencia significativa entre los tiempos promedio de los grupos.")
else:
    print("No podemos rechazar la hipótesis nula: No existe una diferencia significativa entre los tiempos promedio de los grupos.")
```

```python
import pandas as pd

# Crear DataFrame con los resultados
data = {
    'Estadística U': [stat],
    'Valor-p': [p_value],
    'Resultado': ['Rechazamos la hipótesis nula' if p_value < alpha else 'No podemos rechazar la hipótesis nula']
}
df_resultados4 = pd.DataFrame(data)

# Exportar a un archivo CSV
df_resultados4.to_csv('resultados_prueba_mann_whitney_tiempo_total.csv', index=False)

print("Datos exportados a 'resultados_prueba_mann_whitney_tiempo_total.csv'")
```
